```python
from operator import itemgetter

from cog.torque import Graph

from lib.util import load_json

aggregate = load_json('../artifacts/aggregate.json')

g = Graph('sponsorship')

for bill in aggregate:
    state, bill_id, legiscan_bill_id = itemgetter('state', 'bill_id', 'legiscan_bill_id')(bill)
    for sponsor in bill['sponsors']:
        g.put('|'.join([sponsor, state]), 'sponsors', '|'.join([state, bill_id, str(legiscan_bill_id)]))

g.v().inc('sponsors').all()

all_sponsors = set([node['id'] for node in g.v().inc('sponsors').all()['result']])
counts = sorted([
        (sponsor, len(g.v(sponsor).out('sponsors').all()['result']))
        for sponsor
        in all_sponsors
    ],
    key=lambda t: t[1],
    reverse=True
)

counts
```
