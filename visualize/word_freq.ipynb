```python
from glob import glob
from matplotlib import pyplot as plt
import time

from lib.tasks import tokenize_bills
from lib.tasks.visualization import generate_word_cloud
```

```python
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(20, 10))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off")

start_time = time.time()

bill_tokens = tokenize_bills(glob('../bills/*'))
wordcloud_plot = generate_word_cloud(
    bill_tokens,
    '../artifacts/legal_stopwords.json',
    '../configuration/custom_stopwords.json',
)

wordcloud_plot.to_file('../artifacts/cloud.png')
plot_cloud(wordcloud_plot)

end_time = time.time()
print(f'Done after {(end_time-start_time):.2f}s')
```
